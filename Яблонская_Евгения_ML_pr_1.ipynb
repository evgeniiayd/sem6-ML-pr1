```python
# Используем dropna, так как в датасете всего одна пустая строка
X = X.dropna()
print(X.isnull().sum())
```

```python
X.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import RobustScaler, StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.decomposition import PCA

# 1. Стратифицированное разделение на train (70%), val (15%), test (15%)
X_temp, X_test = train_test_split(X, test_size=0.15, stratify=X['quality'], random_state=42)
X_train, X_val = train_test_split(X_temp, test_size=0.1765, stratify=X_temp['quality'], random_state=42)
# 0.1765 ~ 0.15/0.85 для корректного распределения валидационной выборки

# Сохраняем тестовую выборку отдельно для финальной оценки
y_test = X_test['quality']
X_test = X_test.drop(columns=['quality'])
X_test_final = X_test.copy()

y_train = X_train['quality']
X_train = X_train.drop(columns=['quality'])

y_val = X_val['quality']
X_val = X_val.drop(columns=['quality'])

# 2. Проверка пропусков
print("Пропуски в признаках:\n", X_train.isnull().sum())

# только числовые признаки
num_cols = X_train.select_dtypes(include=np.number).columns.tolist()

# 4. Визуализация распределений признаков с помощью boxplot
plt.figure(figsize=(16, 10))
for i, col in enumerate(num_cols):
    plt.subplot(3, 4, i+1)
    sns.boxplot(x=X_train[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# 5. Выявление выбросов методом IQR и удаление (опционально)
# Здесь не удаляем, но можно использовать для анализа
outliers = {}
for col in num_cols:
    Q1 = X_train[col].quantile(0.25)
    Q3 = X_train[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[col] = ((X_train[col] < lower_bound) | (X_train[col] > upper_bound)).sum()
print("Количество выбросов по признакам:\n", outliers)

# 6. Применение RobustScaler для уменьшения влияния выбросов
robust_scaler = RobustScaler()
X_train_robust = X_train.copy()
X_val_robust = X_val.copy()
X_test_robust = X_test_final.copy()

X_train_robust[num_cols] = robust_scaler.fit_transform(X_train[num_cols])
X_val_robust[num_cols] = robust_scaler.transform(X_val[num_cols])
X_test_robust[num_cols] = robust_scaler.transform(X_test_final[num_cols])
```

```python
# 7. Применение StandardScaler для всех числовых признаков
standard_scaler = StandardScaler()
X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test_final.copy()

X_train_scaled[num_cols] = standard_scaler.fit_transform(X_train[num_cols])
X_val_scaled[num_cols] = standard_scaler.transform(X_val[num_cols])
X_test_scaled[num_cols] = standard_scaler.transform(X_test_final[num_cols])

# Сохраняем параметры скейлера для тестовой выборки (уже сохранены в standard_scaler)
```

```python
# 8. Проверка мультиколлинеарности с помощью VIF
vif_data = pd.DataFrame()
vif_data['feature'] = X_train.columns
vif_data['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
print("VIF для признаков:\n", vif_data)

# 9. Создание новых признаков на основе экспертных знаний
# Пример: соотношение fixed_acidity к residual_sugar, ratio_acid_sugar
for df in [X_train_scaled, X_val_scaled, X_test_scaled]:
    df['ratio_acid_sugar'] = df['fixed_acidity'] / (df['residual_sugar'] + 1e-6)  # чтобы избежать деления на 0

# 10. Применение PCA для снижения размерности и визуализации
pca = PCA(n_components=2, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_val_pca = pca.transform(X_val_scaled)
X_test_pca = pca.transform(X_test_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_train_pca[:,0], y=X_train_pca[:,1], hue=y_train, palette='viridis', alpha=0.7)
plt.title('PCA: 2 компоненты')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Quality')
plt.show()
```

Обучающая выборка: X_train_scaled — числовые признаки после заполнения пропусков, применения StandardScaler и добавления новых признаков.
Валидационная выборка: X_val_scaled — аналогично обработанная, для настройки и проверки модели.
Тестовая выборка: X_test_scaled — отдельно сохранённая и обработанная таким же образом, для финальной оценки модели.

# Обучение машинных моделей и нахождение наиболее удачной

```python
# Проверьте тип целевой переменной
print(y_train.value_counts())  # Для проверки уникальных значений
print(y_val.value_counts())
print(y_test.value_counts())
```

```python
import numpy as np
import pandas as pd
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.metrics import f1_score, accuracy_score, precision_recall_fscore_support, roc_auc_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Предполагается, что X_train_scaled, y_train, X_val_scaled, y_val, X_test_scaled, y_test уже определены
# y_train, y_val, y_test — целевые переменные, например: y_train = X_train_scaled['quality']

# 1. Baseline модели
models = {
    'Dummy_MostFrequent': DummyClassifier(strategy='most_frequent', random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
}

# 2. Продвинутые классификаторы с базовыми параметрами
models.update({
    'RandomForest': RandomForestClassifier(random_state=42),
    'SVM_rbf': SVC(kernel='rbf', probability=True, random_state=42),
    'KNN': KNeighborsClassifier(),
    'MLP': MLPClassifier(max_iter=1000, random_state=42),
})

# 3. Обучение базовых моделей и оценка на валидации
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    y_proba = model.predict_proba(X_val) if hasattr(model, "predict_proba") else None

    acc = accuracy_score(y_val, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(y_val, y_pred, average='weighted')
    roc_auc = roc_auc_score(y_val, y_proba, multi_class='ovr') if y_proba is not None else np.nan
    cm = confusion_matrix(y_val, y_pred)

    results[name] = {
        'model': model,
        'accuracy': acc,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'roc_auc': roc_auc,
        'confusion_matrix': cm
    }
    print(f"Model: {name}")
    print(f"Accuracy: {acc:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}, F1: {f1:.4f}, ROC AUC: {roc_auc:.4f}")
    print("Confusion Matrix:\n", cm)
    print("-"*40)

# 4. Пример гиперпараметрического тюнинга для RandomForest с RandomizedSearchCV
from scipy.stats import randint

param_dist_rf = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(3, 20),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
    'bootstrap': [True, False]
}

rf = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(rf, param_distributions=param_dist_rf, n_iter=20, cv=3, scoring='f1_weighted', random_state=42, n_jobs=-1)
random_search.fit(X_train, y_train)

print("Best RF params:", random_search.best_params_)
best_rf = random_search.best_estimator_

# Оценка лучшей RF на валидации
y_pred_rf = best_rf.predict(X_val)
y_proba_rf = best_rf.predict_proba(X_val)
print("Tuned RF Validation F1:", f1_score(y_val, y_pred_rf, average='weighted'))
print("Tuned RF Validation ROC AUC:", roc_auc_score(y_val, y_proba_rf, multi_class='ovr'))

# 5. Обучение ансамбля VotingClassifier (логистическая регрессия, дерево решений, RF)
voting_clf = VotingClassifier(
    estimators=[
        ('lr', models['LogisticRegression']),
        ('dt', models['DecisionTree']),
        ('rf', best_rf)
    ],
    voting='soft'
)
voting_clf.fit(X_train, y_train)
y_pred_vote = voting_clf.predict(X_val)
y_proba_vote = voting_clf.predict_proba(X_val)
print("Voting Classifier F1:", f1_score(y_val, y_pred_vote, average='weighted'))
print("Voting Classifier ROC AUC:", roc_auc_score(y_val, y_proba_vote, multi_class='ovr'))

# 6. Финальная оценка на тестовой выборке лучшей модели (например, tuned RF или Voting)
best_model = best_rf  # или voting_clf, в зависимости от результатов

y_test_pred = best_model.predict(X_test)
y_test_proba = best_model.predict_proba(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Test Precision, Recall, F1:", precision_recall_fscore_support(y_test, y_test_pred, average='weighted')[:3])
print("Test ROC AUC:", roc_auc_score(y_test, y_test_proba, multi_class='ovr'))
print("Test Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Классы
classes = [3, 4, 5, 6, 7, 8, 9]
n_classes = len(classes)

# Бинаризуем метки
y_test_bin = label_binarize(y_test, classes=classes)

# y_test_proba — вероятности для всех классов, shape (n_samples, n_classes)
# Убедитесь, что y_test_proba соответствует этому порядку классов (3,4,...,9)

fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_test_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10,8))
colors = plt.cm.get_cmap('tab10', n_classes)

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors(i),
             label=f'Class {classes[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0,1], [0,1], 'k--', label='Random guess')
plt.xlim([0,1])
plt.ylim([0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Multiclass Classification (One-vs-Rest)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

Лучшей моделью из представленных является Random Forest с тюнингом гиперпараметров. Почему:

Имеет наилучшее качество по ключевым метрикам: F1 на валидации ~0.66 (лучше, чем у базовых моделей и ансамбля Voting),
Высокий ROC AUC (0.77 на валидации, 0.82 на тесте) — хороший баланс между чувствительностью и специфичностью,
Accuracy и Precision выше, чем у большинства моделей,
Более стабильные и интерпретируемые результаты по сравнению с MLP и SVM,
Тюнинг гиперпараметров улучшил модель по сравнению с базовым RF.
